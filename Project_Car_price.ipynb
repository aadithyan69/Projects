```python
#get_dummies encoding
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df1
```

```python
#concatenation(include remaining data with the new df1)
dfc=pd.concat([df,df1],axis=1) #concat df and df1
dfc
```

```python
#drop unwanted columns
#
dfc=dfc.drop(['Unnamed: 0','Name','New_Price','Location','Fuel_Type','Transmission','Owner_Type','Fuel_Type_Electric'],axis=1)
dfc
```

```python
#replacing the units of the following columns(kmpl,CC,bhp etc),
#'null' is also replace with 0.
dfc['Mileage']=dfc['Mileage'].str.replace('km/kg','')
dfc['Mileage']=dfc['Mileage'].str.replace('kmpl','')
dfc['Engine']=dfc['Engine'].str.replace('CC','')
dfc['Power']=dfc['Power'].str.replace('bhp','')
dfc['Mileage']=dfc['Mileage'].str.replace('null','0')
dfc['Engine']=dfc['Engine'].str.replace('null','0')
dfc['Power']=dfc['Power'].str.replace('null','0')
dfc
```

```python
dfc.dtypes
```

```python
#Data type conversion
#Mileage=float
#Engine=float
#Power=float
dfc['Mileage']=dfc['Mileage'].astype(float)
dfc['Engine']=dfc['Engine'].astype(float)
dfc['Power']=dfc['Power'].astype(float)
```

```python
#in above steps we replaced null with 0,here we are making zero as a missing value,
#because zero cannot a value for mileage,engine,power practically.
dfc.loc[dfc.Mileage==0,'Mileage']=np.NaN
dfc.loc[dfc.Power==0,'Power']=np.NaN
dfc.loc[dfc.Engine==0,'Engine']=np.NaN
```

```python
dfc.isna().sum()
```

```python
dfc['Mileage']=dfc['Mileage'].fillna(dfc['Mileage'].mean())
dfc['Engine']=dfc['Engine'].fillna(dfc['Engine'].mean())
dfc['Power']=dfc['Power'].fillna(dfc['Power'].mean())
dfc['Seats']=dfc['Seats'].fillna(dfc['Seats'].mode()[0])
```

```python
dfc.isna().sum()
```

```python
x=dfc.drop(['Price'],axis=1)
y=dfc['Price']
#preprocessing in training data is completed here.
```
